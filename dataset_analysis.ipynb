```python
import pandas as pd

def merge_csv_files(csv_files, output_csv):
    """
    Junta mltiplos arquivos CSV em um nico arquivo.

    Args:
        csv_files (list): Lista de caminhos para os arquivos CSV a serem unidos.
        output_csv (str): Caminho para o arquivo CSV de sada.
    """
    try:
        # Ler todos os CSVs em dataframes e concaten-los
        dataframes = [pd.read_csv(csv_file) for csv_file in csv_files]
        merged_dataframe = pd.concat(dataframes, ignore_index=True)

        # Salvar o dataframe resultante em um nico CSV
        merged_dataframe.to_csv(output_csv, index=False)

        print(f"CSV combinado gerado com sucesso: {output_csv}")

    except Exception as e:
        print(f"Erro ao juntar os arquivos CSV: {e}")

# Exemplo de uso
merge_csv_files(
    csv_files=[
        "/home/carlos/Documents/pessoal/case/dataset/output_open.csv",
        "/home/carlos/Documents/pessoal/case/dataset/output_closed.csv"
    ],
    output_csv="/home/carlos/Documents/pessoal/case/dataset/merged_output.csv"
)
```

```python
import os
import shutil
import random

# Diretrios de origem
input_dir_closed = "/home/carlos/Documents/pessoal/case/dataset/input/closed"
input_dir_open = "/home/carlos/Documents/pessoal/case/dataset/input/open"

# Diretrios de destino para treinamento e validao
output_dir_train_closed = "/home/carlos/Documents/pessoal/case/dataset/small_dataset/train/closed"
output_dir_val_closed = "/home/carlos/Documents/pessoal/case/dataset/small_dataset/validation/closed"
output_dir_train_open = "/home/carlos/Documents/pessoal/case/dataset/small_dataset/train/open"
output_dir_val_open = "/home/carlos/Documents/pessoal/case/dataset/small_dataset/validation/open"

# Quantidade de imagens para selecionar
num_images = 999999  # Ajuste conforme necessrio

# Proporo de diviso entre treinamento e validao
train_ratio = 0.8  # 80% para treinamento, 20% para validao

# Funo para criar diretrios, se no existirem
def create_dirs(*dirs):
    for directory in dirs:
        os.makedirs(directory, exist_ok=True)

# Funo para selecionar, dividir e copiar imagens
def select_divide_and_copy_images(input_dir, output_dir_train, output_dir_val, num_images, train_ratio):
    # Listar todas as imagens no diretrio de origem
    images = [img for img in os.listdir(input_dir) if img.lower().endswith(('jpg', 'jpeg', 'png'))]
    
    # Embaralhar as imagens e selecionar as primeiras 'num_images'
    selected_images = random.sample(images, min(num_images, len(images)))
    
    # Calcular o nmero de imagens para treinamento e validao
    num_train = int(len(selected_images) * train_ratio)
    
    # Dividir as imagens
    train_images = selected_images[:num_train]
    val_images = selected_images[num_train:]
    
    # Copiar imagens de treinamento
    for image in train_images:
        src_path = os.path.join(input_dir, image)
        dst_path = os.path.join(output_dir_train, image)
        shutil.copy(src_path, dst_path)
        #print(f"Copiado para treinamento: {image} -> {dst_path}")
    
    # Copiar imagens de validao
    for image in val_images:
        src_path = os.path.join(input_dir, image)
        dst_path = os.path.join(output_dir_val, image)
        shutil.copy(src_path, dst_path)
        #print(f"Copiado para validao: {image} -> {dst_path}")

# Criar diretrios de destino
create_dirs(output_dir_train_closed, output_dir_val_closed, output_dir_train_open, output_dir_val_open)

# Selecionar, dividir e copiar imagens para a classe 'closed'
select_divide_and_copy_images(input_dir_closed, output_dir_train_closed, output_dir_val_closed, num_images, train_ratio)

# Selecionar, dividir e copiar imagens para a classe 'open'
select_divide_and_copy_images(input_dir_open, output_dir_train_open, output_dir_val_open, num_images, train_ratio)

print("Seleo, diviso e cpia concludas!")
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping
import matplotlib.pyplot as plt

# Caminho dos dados
train_dir = '/home/carlos/Documents/pessoal/case/dataset/small_dataset/train'
val_dir = '/home/carlos/Documents/pessoal/case/dataset/small_dataset/validation'

# Configurao de dados com Data Augmentation para treino
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=40,       # Rotao aleatria de 40 graus
    width_shift_range=0.2,   # Deslocamento horizontal
    height_shift_range=0.2,  # Deslocamento vertical
    shear_range=0.2,         # Cisalhamento
    zoom_range=0.3,          # Zoom aleatrio
    horizontal_flip=True,    # Flip horizontal
    fill_mode='nearest'      # Preenchimento de pixels vazios
)

# Apenas rescaling para validao
val_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(150, 150),
    batch_size=32,
    class_mode='binary'
)

val_generator = val_datagen.flow_from_directory(
    val_dir,
    target_size=(150, 150),
    batch_size=32,
    class_mode='binary'
)

# Modelo
from tensorflow.keras import regularizers

model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(32, (3,3), activation='relu', input_shape=(150, 150, 3)),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(128, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu', kernel_regularizer=regularizers.l2(0.01)),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(64, activation='relu', kernel_regularizer=regularizers.l2(0.01)),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(1, activation='sigmoid')  # Sada binria
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Callbacks: ReduceLROnPlateau e EarlyStopping
reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=3,
    min_lr=1e-5
)

early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True
)

# Treinamento
history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // train_generator.batch_size,
    validation_data=val_generator,
    validation_steps=val_generator.samples // val_generator.batch_size,
    epochs=50,  # Aumentei o nmero de pocas
    callbacks=[reduce_lr, early_stopping]
)

# Salvar modelo
model.save('/home/carlos/Documents/pessoal/case/model/garrafa_classifier.h5')

# Plotar resultados
plt.plot(history.history['accuracy'], label='Acurcia Treino')
plt.plot(history.history['val_accuracy'], label='Acurcia Validao')
plt.xlabel('pocas')
plt.ylabel('Acurcia')
plt.legend()
plt.show()
```

```python
import os
import shutil
import random

# Diretrios iniciais
closed_images_dir = '/home/carlos/Documents/pessoal/case/dataset/yolo_dataset/closed_images_'  # Pasta A
closed_labels_dir = '/home/carlos/Documents/pessoal/case/dataset/yolo_dataset/close_labels'  # Pasta B
open_labels_dir = '/home/carlos/Documents/pessoal/case/dataset/yolo_dataset/labels_open_'      # Pasta C
open_images_dir = '/home/carlos/Documents/pessoal/case/dataset/yolo_dataset/open_images'    # Pasta D

# Diretrio de sada
output_dir = '/home/carlos/Documents/pessoal/case/dataset/yolo_dataset/'
train_images_dir = os.path.join(output_dir, 'train/images')
train_labels_dir = os.path.join(output_dir, 'train/labels')
val_images_dir = os.path.join(output_dir, 'val/images')
val_labels_dir = os.path.join(output_dir, 'val/labels')

# Criar diretrios
os.makedirs(train_images_dir, exist_ok=True)
os.makedirs(train_labels_dir, exist_ok=True)
os.makedirs(val_images_dir, exist_ok=True)
os.makedirs(val_labels_dir, exist_ok=True)

# Configurar proporo de treino e validao
train_ratio = 0.8  # 80% treino, 20% validao

# Coletar dados "open" (imagens e rtulos)
open_data = []
for txt_file in os.listdir(open_labels_dir):
    if txt_file.endswith('.txt'):
        image_name = os.path.splitext(txt_file)[0] + '.jpg'
        image_path = os.path.join(open_images_dir, image_name)
        label_path = os.path.join(open_labels_dir, txt_file)
        if os.path.exists(image_path):
            open_data.append((image_path, label_path))

# Coletar dados "closed" (imagens e rtulos)
closed_data = []
for txt_file in os.listdir(closed_labels_dir):
    if txt_file.endswith('.txt'):
        image_name = os.path.splitext(txt_file)[0] + '.jpg'
        image_path = os.path.join(closed_images_dir, image_name)
        label_path = os.path.join(closed_labels_dir, txt_file)
        if os.path.exists(image_path):
            closed_data.append((image_path, label_path))

# Combinar e embaralhar os dados
all_data = open_data + closed_data
random.shuffle(all_data)

# Dividir entre treino e validao
num_train = int(len(all_data) * train_ratio)
train_data = all_data[:num_train]
val_data = all_data[num_train:]

# Funo para copiar arquivos
def copy_files(data, images_dir, labels_dir):
    for image_path, label_path in data:
        # Copiar imagem
        shutil.copy(image_path, os.path.join(images_dir, os.path.basename(image_path)))
        # Copiar rtulo
        shutil.copy(label_path, os.path.join(labels_dir, os.path.basename(label_path)))

# Copiar arquivos para treino e validao
copy_files(train_data, train_images_dir, train_labels_dir)
copy_files(val_data, val_images_dir, val_labels_dir)

print("Organizao concluda!")
```

```python
import os
from ultralytics import YOLO

# Diretrios do dataset
train_images_dir = '/home/carlos/Documents/pessoal/case/dataset/yolo_dataset/train/images'
val_images_dir = '/home/carlos/Documents/pessoal/case/dataset/yolo_dataset/val/images'

# Caminho do arquivo YAML
data_yaml_path = '/home/carlos/Documents/pessoal/case/dataset/yolo_dataset/data.yaml'

# Criar o arquivo YAML dinamicamente
data_yaml_content = f"""
train: {train_images_dir}
val: {val_images_dir}

nc: 2
names: ['open', 'closed']
"""
with open(data_yaml_path, 'w') as yaml_file:
    yaml_file.write(data_yaml_content)

print(f"Arquivo data.yaml criado em: {data_yaml_path}")

# Treinamento usando YOLOv8
model = YOLO('yolov8n.pt')  # Use 'yolov8s.pt' para um modelo maior

results = model.train(
    data=data_yaml_path,  # Caminho para o arquivo YAML
    epochs=50,            # Nmero de pocas
    imgsz=640,            # Tamanho da imagem
    batch=16,             # Tamanho do batch
    device=0              # GPU (0) ou CPU (-1)
)

# Avaliao
metrics = model.val()

# Verificar o diretrio para inferncia
predict_dir = '/home/carlos/Documents/pessoal/case/dataset/yolo_dataset/val/images'
supported_formats = {'jpg', 'jpeg', 'png', 'bmp', 'webp', 'mp4', 'avi', 'mkv'}

files = [f for f in os.listdir(predict_dir) if f.split('.')[-1] in supported_formats]
if not files:
    raise FileNotFoundError(f"Nenhum arquivo suportado encontrado no diretrio: {predict_dir}")
else:
    print(f"Arquivos para inferncia encontrados: {files}")

# Inferncia
results = model.predict(source=predict_dir, save=True)
```
